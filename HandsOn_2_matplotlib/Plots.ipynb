```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

my_data = pd.read_csv("/home/mldantas/Dropbox/IIPBC/Vostok-tpt-co2.csv")

temperature    = my_data["Temp"]
carbon_dioxide = my_data["CO2"]
radiation      = my_data["Radiation"]

plt.plot(temperature, carbon_dioxide, 'o')
plt.xlabel("Temperature", fontsize=15)
plt.ylabel("Carbon Dioxide", fontsize=15)
plt.show()
```

```python
plt.semilogy(temperature, carbon_dioxide, 'o')
plt.xlabel("Temperature", fontsize=15)
plt.ylabel("Carbon Dioxide", fontsize=15)
plt.show()
```

```python
plt.hist(temperature, facecolor='red')
plt.show()
```

```python
plt.hist(temperature, 50, color="#b3ffb3") #html color code
plt.xlabel("Temperature", fontsize=15)
plt.ylabel("Amount", fontsize=15)
plt.tick_params('both', labelsize=30)
plt.show()
```

```python
matplotlib.style.use("ggplot")
```

```python
plt.hist(temperature, 50, color="red", alpha=0.6) #html color code
plt.text(-6, 200, "This is the coolest histogram in History!", fontsize=12)
plt.xlabel("Temperature", fontsize=15)
plt.ylabel("Amount", fontsize=15)
plt.tick_params('both', labelsize=10)
plt.show()
```

```python
plt.scatter(temperature, carbon_dioxide, color="#ff0000", alpha=0.4) #html color code
plt.xlabel(r"Temperature ($^o$C)", fontsize=15)
plt.ylabel(r"$CO_2$", fontsize=15)
plt.tick_params('both', labelsize=10)
plt.grid(alpha=0.5)
plt.show()
```
